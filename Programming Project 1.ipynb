```python
def Rock_Paper_Scissors():
    """Plays a sequence of Rock, Paper, Scissors games with the user against the computer.""" 
    # identifiers are used to keep a running track of games played, number of games won by the user, 
    # number of games won by the computer, and the number of tied games.
    games_played = 0
    user_score = 0
    computer_score = 0
    tied_games = 0
    
    while True: 
        # While loop allows for multiple iterations of the Rock_Paper_Scissors() function
        
        # Computer move is based on the computer randomly selecting rock, paper or scissors
        # Random module and choice function is needed in order for the computer to select its move
        import random
        possible_moves = ["rock", "paper", "scissors"]
        computer_move = random.choice(possible_moves)
        
        # User move is input based for each game.
        # Note: The code uses user_move.lower() to refer back to the user input,
        # this is to make the program more user-friendly.
        # User can break the loop by typing finish when prompted with their move.
        user_move = input("What's your move? Rock, paper or scissors. (Type 'Finish' to stop playing): ")
        
        # Depening on the Users inputted move, the following lines will be printed.
        if user_move.lower() != "finish":
            print("Rock, Paper, Scissors, Go...")
        else:
            print("You chose to stop playing.")
        
        
        # If-Elif-Else block for the possible combination of user-computer moves
        if user_move.lower() == computer_move:  
            # Condition for when the user and computer play the same move
            # Playing the same move reults in that game being a tie. No winner. 
            games_played += 1
            tied_games += 1
            print(f"Result: Tie. \nReason: You played {user_move.lower()}, computer played {computer_move}.")
        
        elif user_move.lower() == "rock": # Condition for when the user plays rock
            if computer_move == "scissors": 
                # Nested condition for when the user plays rock and computer plays scissors
                # Results in a user win. Adds 1 to the running total of games played and user wins.
                games_played += 1
                user_score += 1
                print(f"Result: You Won!. \nReason: You played {user_move.lower()}, computer played {computer_move}. Rock smashes scissors.")
            else:
                # Nested condition for when the user plays rock and computer plays paper
                # Results in a computer win. Adds 1 to the running total of games played and computer wins.
                games_played += 1
                computer_score += 1
                print(f"Result: You Lost. \nReason: You played {user_move.lower()}, computer played {computer_move}. Paper covers rock.")
        
        
        elif user_move.lower() == "paper": # Condition for when the user plays paper
            if computer_move == "rock":
                # Nested condition for when the user plays paper and computer plays rock
                # Results in a user win. Adds 1 to the running total of games played and user wins.
                games_played += 1
                user_score += 1
                print(f"Result: You Won! \nReason: You played {user_move.lower()}, computer played {computer_move}. Paper covers rock.")
            else:  
                # Nested condition for when the user plays paper and computer plays scissors
                # Results in a computer win. Adds 1 to the running total of games played and computer wins.
                games_played += 1
                computer_score += 1
                print(f"Result: You Lost. \nReason: You played {user_move.lower()}, computer played {computer_move}. Scissors cut paper.")
        
    
        elif user_move.lower() == "scissors": # Conditional statement if the user plays scissors.
            if computer_move == "paper": 
                # Nested condition for when the user plays scissors and computer plays paper. 
                # Results in a user win. Adds 1 to the running total of games played and user wins.
                games_played += 1
                user_score += 1
                print(f"Result: You Won! \nReason: You played {user_move.lower()}, computer played {computer_move}. Scissors cut paper.")
            else: 
                # Nested condition for when the user plays scissors and computer plays rock, 
                # Results in a computer win. Adds 1 to the running total of games played and computer wins.
                games_played += 1
                computer_score += 1
                print(f"Result: You Lost. \nReason: You played {user_move.lower()}, computer played {computer_move}. Rock smashes scissors.")
          
        
        elif user_move.lower() == "finish": 
            # If the user input is 'finish' then the while loop will break.
            # Has no effect on the number of games played or scores of either the user or computer
            break
        
        
        else: 
            # Condition if user input doesn't meet one of the 3 possible user moves.
            print(f"Check spelling. '{user_move}' is either misspelled or an invalid move.")
            
        print("")
    
    
    print("")
    print("**********************************************Scoreboard**********************************************")
    if user_score == computer_score:
        # Condition that returns the following scoreboard lines
        # for when the user and computer win the same amount of games.
        print(f"In {games_played} games played, no overall winner. You and the computer each won the same amount of games.")
        print(f"\t Number of games won by the user: {user_score}")
        print(f"\t Number of garockmes won by the computer: {computer_score}")
        print(f"\t Number of games resulting in a tie: {tied_games}")

    elif user_score > computer_score:
        # Condition that returns the following scoreboard lines
        # for when the user wins more games than the computer.
        print(f"In {games_played} games played, you won more overall games than the computer.")
        print(f"\t Number of games won by the user: {user_score}")
        print(f"\t Number of games won by the computer: {computer_score}")
        print(f"\t Number of games resulting in a tie: {tied_games}")

    else:
        # Condition that returns the following scoreboard lines
        # for when the computer wins more games than the user.
        print(f"In {games_played} games played, the computer won more overall games than you.")
        print(f"\t Number of games won by the user: {user_score}")
        print(f"\t Number of games won by the computer: {computer_score}")
        print(f"\t Number of games resulting in a tie: {tied_games}")

    print("******************************************************************************************************")
```

```python
Rock_Paper_Scissors()
```
